```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris_data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', 
                        header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
```

```python
# 1. List down the features and their types available in the dataset
print('Features and their types:')
print(iris_data.dtypes)
```

```python
# 2. Create a histogram for each feature in the dataset
iris_data.hist(figsize=(10, 8))
plt.show()
```

```python
# 3. Create a boxplot for each feature in the dataset
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
for ax, column in zip(axes.flatten(), iris_data.columns[:-1]):
    sns.boxplot(x='class', y=column, data=iris_data, ax=ax)
plt.tight_layout()
plt.show()
```
